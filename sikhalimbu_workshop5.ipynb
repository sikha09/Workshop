```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Step 1: Define the function and its gradient
def f(x):
    return x**2

def grad_f(x):
    return 2 * x

# Step 2: Initialize parameters
x_start = 10  # Starting point
learning_rate = 0.1
iterations = 50

# Step 3: Gradient Descent Loop
x_values = [x_start]  # Store x values for visualization
f_values = [f(x_start)]  # Store function values for visualization

x = x_start
for i in range(iterations):
    gradient = grad_f(x)
    x = x - learning_rate * gradient
    x_values.append(x)
    f_values.append(f(x))

# Step 4: Visualization
x_plot = np.linspace(-10, 10, 500)  # Values for plotting f(x)
plt.figure(figsize=(10, 6))

# Plot the function
plt.plot(x_plot, f(x_plot), label="f(x) = x^2", color="blue")

# Plot the gradient descent steps
plt.scatter(x_values, f_values, color="red", label="Gradient Descent Steps")
for i in range(len(x_values) - 1):
    plt.arrow(x_values[i], f_values[i], x_values[i + 1] - x_values[i], f_values[i + 1] - f_values[i],
              color='orange', head_width=0.3, head_length=0.3)

plt.title("Gradient Descent to Minimize f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 1: Define the cubic function and its gradient
def f(x):
    return x**3 - 4 * x**2 + 6 * x

def grad_f(x):
    return 3 * x**2 - 8 * x + 6

# Step 2: Initialize parameters
x_start = 0  # Starting point
learning_rate = 0.1
threshold = 0.0001
max_iterations = 5

# Step 3: Gradient Descent Loop
x_values = [x_start]  # Store x values for visualization
f_values = [f(x_start)]  # Store function values for visualization

x_old = x_start
for i in range(max_iterations):
    gradient = grad_f(x_old)
    x_new = x_old - learning_rate * gradient
    x_values.append(x_new)
    f_values.append(f(x_new))

    if abs(x_new - x_old) < threshold:
        break

    x_old = x_new

# Step 4: Visualization
x_plot = np.linspace(-1, 3, 500)  # Values for plotting f(x)
plt.figure(figsize=(10, 6))

# Plot the function
plt.plot(x_plot, f(x_plot), label="f(x) = x^3 - 4x^2 + 6x", color="blue")

# Plot the gradient descent steps
plt.scatter(x_values, f_values, color="red", label="Gradient Descent Steps")
for i in range(len(x_values) - 1):
    plt.arrow(x_values[i], f_values[i], x_values[i + 1] - x_values[i], f_values[i + 1] - f_values[i],
              color='orange', head_width=0.1, head_length=0.2)

plt.title("Gradient Descent to Minimize f(x) = x^3 - 4x^2 + 6x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 1: Define the cost function for linear regression and its gradient
def compute_cost(w, b, X, y):
    m = X.shape[0]
    predictions = X.dot(w) + b
    errors = predictions - y
    return (1 / (2 * m)) * np.sum(errors**2)

def compute_gradient(w, b, X, y):
    m = X.shape[0]
    predictions = X.dot(w) + b
    errors = predictions - y
    grad_w = (1 / m) * X.T.dot(errors)
    grad_b = (1 / m) * np.sum(errors)
    return grad_w, grad_b

# Step 2: Initialize parameters
n_features = 2  # Example: number of features
w = np.zeros(n_features)  # Initialize weights
b = 0  # Initialize bias
learning_rate = 0.01
threshold = 1e-6
max_iterations = 1000

# Example data (X, y)
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])  # Features
y = np.array([5, 7, 9, 11])  # Labels

# Step 3: Gradient Descent Loop
cost_history = []
for i in range(max_iterations):
    cost = compute_cost(w, b, X, y)
    cost_history.append(cost)

    grad_w, grad_b = compute_gradient(w, b, X, y)

    w -= learning_rate * grad_w
    b -= learning_rate * grad_b

    if i > 0 and abs(cost_history[-1] - cost_history[-2]) < threshold:
        break

# Step 4: Visualization
plt.figure(figsize=(10, 6))

# Plot cost function over iterations
plt.plot(cost_history, label="Cost Function", color="blue")
plt.title("Gradient Descent to Minimize MSE for Linear Regression")
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")
plt.legend()
plt.grid()
plt.show()

# Print final weights and bias
print(f"Final weights: {w}")
print(f"Final bias: {b}")
```
